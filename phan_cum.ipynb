```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# =========================
# ĐỌC FILE EXCEL
# =========================
df = pd.read_excel(
    "TỔNG HỢP ĐIỂM K58KTP (1).xlsx",
    header=None
)

students = []

# =========================
# DUYỆT TỪ CỘT D TRỞ ĐI
# =========================
for col in range(3, df.shape[1]):

    try:

        # MSSV ở hàng 1
        mssv = str(df.iloc[1, col]).strip()

        # Họ tên ở hàng 2
        name = str(df.iloc[2, col]).strip()

        # bỏ cột rỗng
        if mssv == "nan" or name == "nan":
            continue

        # =========================
        # LẤY ĐIỂM
        # =========================
        scores = pd.to_numeric(
            df.iloc[4:, col],
            errors="coerce"
        )

        # bỏ NaN
        scores = scores.dropna()

        # chỉ lấy điểm từ 0 -> 4
        scores = scores[
            (scores >= 0) &
            (scores <= 4)
        ]

        # nếu không có điểm
        if len(scores) == 0:
            continue

        # tính điểm trung bình
        avg_score = round(scores.mean(), 2)

        # thêm sinh viên
        students.append({
            "MSSV": mssv,
            "Họ tên": name,
            "Điểm TB": avg_score
        })

    except:
        continue

# =========================
# TẠO DATAFRAME
# =========================
student_df = pd.DataFrame(students)

print("Tổng sinh viên:", len(student_df))

# =========================
# PHÂN CỤM KMEANS
# =========================
X = student_df[["Điểm TB"]]

# chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

student_df["Cluster"] = kmeans.fit_predict(X_scaled)

# =========================
# XÁC ĐỊNH HỌC LỰC
# =========================
cluster_mean = student_df.groupby("Cluster")["Điểm TB"].mean()

sorted_cluster = cluster_mean.sort_values()

low_cluster = sorted_cluster.index[0]
mid_cluster = sorted_cluster.index[1]
high_cluster = sorted_cluster.index[2]

# đổi tên nhóm
mapping = {
    low_cluster: "Học lực thấp",
    mid_cluster: "Học lực trung bình",
    high_cluster: "Học lực cao"
}

student_df["Nhóm học lực"] = student_df["Cluster"].map(mapping)

# xóa cột cluster
student_df.drop(columns=["Cluster"], inplace=True)

# =========================
# SẮP XẾP THEO ĐIỂM
# =========================
student_df = student_df.sort_values(
    by="Điểm TB",
    ascending=False
)

# reset index
student_df.reset_index(drop=True, inplace=True)

# =========================
# HIỂN THỊ
# =========================
print("\nDANH SÁCH PHÂN CỤM:")
print(student_df)

# =========================
# THỐNG KÊ
# =========================
print("\nTHỐNG KÊ NHÓM:")
print(student_df["Nhóm học lực"].value_counts())

# =========================
# XUẤT FILE
# =========================
student_df.to_excel(
    "phan_cum_hoc_luc.xlsx",
    index=False
)

print("\nĐã phân cụm xong!")
print("Đã xuất file: phan_cum_hoc_luc.xlsx")
```
